**Univariate Analysis**

Margin Captilisation looks like?

```python
sns.distplot(df['Mar Cap - Crore'])
plt.axvline(x=np.mean(df['Mar Cap - Crore']),c='Red',ls='--',label='mean')
plt.axvline(x=np.percentile(df['Mar Cap - Crore'],25),c='green',ls='--',label='25-75th percentile')
plt.axvline(x=np.percentile(df['Mar Cap - Crore'],25),c='green',ls='--')
plt.legend()
```

```python
df.hist(figsize=(10,10))
```

Sales Qtr looks like?

```python
sns.distplot(df['Sales Qtr - Crore'])
plt.axvline(x=np.mean(df['Sales Qtr - Crore']),c='Red',ls='--',label='mean')
plt.axvline(x=np.percentile(df['Sales Qtr - Crore'],25),c='green',ls='--',label='25-75th percentile')
plt.axvline(x=np.percentile(df['Sales Qtr - Crore'],25),c='green',ls='--')
plt.legend()
```
